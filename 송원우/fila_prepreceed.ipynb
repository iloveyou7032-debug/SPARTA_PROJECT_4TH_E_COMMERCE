```python
# 1. 파일 경로 설정
product_file = './data/fila_product_master_260420.csv'
review_file = './data/fila_reviews_master_260420.csv'

# 2. 데이터 로드
df_products = pd.read_csv(product_file)
df_reviews = pd.read_csv(review_file)

# 3. 리뷰 파일의 'search_keyword'(상품 ID)를 카운트하여 집계
review_counts = df_reviews['search_keyword'].value_counts().reset_index()
review_counts.columns = ['product_id', 'new_count']

# 4. 상품 데이터와 병합하여 리뷰 개수 업데이트
# 기존에 review_count 컬럼이 있더라도 새로운 집계값으로 덮어씌웁니다.
df_final = pd.merge(df_products, review_counts, on='product_id', how='left')
df_final['review_count'] = df_final['new_count'].fillna(0).astype(int)

# 5. 불필요한 임시 컬럼 삭제
df_final = df_final.drop(columns=['new_count'])

# 6. 원본 파일명 그대로 저장 (덮어쓰기)
df_final.to_csv(product_file, index=False, encoding='utf-8-sig')

print(f"✅ {product_file} 파일에 실제 리뷰 개수 업데이트가 완료되었습니다.")
```
