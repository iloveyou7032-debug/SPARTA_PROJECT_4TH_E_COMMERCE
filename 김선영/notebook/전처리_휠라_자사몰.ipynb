## purchase_option 결측치 처리

```python
df['purchase_option_color'] = df['purchase_option_color'].astype(object)
df['purchase_option_size'] = df['purchase_option_size'].astype(object)
```

```python
# 1)
# df[df['product_name'] == '러버스 트라이앵글 브라']['source_url'].value_counts(dropna=False)
# 동일 제품명에 url 두 가지 -> 제품명 동일하나 url 별로 컬러구성이 다름 & 사이즈 구성은 동일함.
# 085, 090, 095, 100 / 85(XS), 90(S), 95(M), 100(L) ---> 숫자로 통일
import re

def normalize_size_value(text):
    if pd.isna(text):
        return None
    
    text = str(text).strip()
    
    # 괄호 및 괄호 안 내용 제거 (90(S) → 90)
    text = re.sub(r'\(.*?\)', '', text).strip()
    
    # 숫자만 있는 경우
    if re.fullmatch(r'\d+', text):
        num = int(text)
        # 사이즈 범위(60~320)인 경우만 앞의 0 제거
        if 60 <= num <= 320:
            return str(num)
        else:
            return text  # 범위 밖이면 원본 유지
    
    return text

mask = df['product_name'] == '러버스 트라이앵글 브라'

df.loc[mask, 'purchase_option_size'] = (
    df.loc[mask, 'purchase_option']
    .apply(normalize_size_value)
)

df[mask][['purchase_option', 'purchase_option_size']].value_counts()
```

```python
# 2) 푸퍼 미니 토트백
# url은 한 종류로 동일, product_code 등도 한 종류 -> 화이트/블랙 구별 불가 -> purchase_option_color 컬럼 채우기 패스
# purchase_option 종류: 001, 000(free), 150 -> url 확인 시 모드 동일 제품으로 이동함.
#                       -> 옵션에 사이즈 모두 '000(free)'만 있음.
#                       -> 리뷰 직접 확인해보니 '001 = 블랙', '150 = 화이트', '217 = 브라운'
# purchase_option_size 모두 000으로 처리
# purchase_option_color 컬러 1:1 매칭
mask = df['product_name'] == '푸퍼 미니 토트백'

# purchase_option 값이 있는 데이터의 purchase_option_size를 '000'으로
df.loc[mask & df['purchase_option'].notna(), 'purchase_option_size'] = '000'

# purchase_option 값에 따라 purchase_option_color 채우기
df.loc[mask & (df['purchase_option'] == '001'), 'purchase_option_color'] = '블랙'
df.loc[mask & (df['purchase_option'] == '150'), 'purchase_option_color'] = '화이트'

# 확인
df[mask][['purchase_option', 'purchase_option_color', 'purchase_option_size']].value_counts(dropna=False)
```

```python
# ========================
# 최종 컬럼 정리
# ========================

# 1) 데이터 형변환
def convert_dtypes(df):
    # 날짜 컬럼
    for col in ["collect_date", "review_date"]:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    # ID 컬럼
    for col in ["review_id", "product_id"]:
        if col in df.columns:
            df[col] = df[col].astype("string")

    # 문자열 컬럼
    for col in ["product_name", "content", "purchase_option", "product_url"]:
        if col in df.columns:
            df[col] = df[col].astype("string")

    # category 컬럼
    for col in ["platform", "purchase_option_color", "purchase_option_size", "user_height_group", "user_weight_group"]:
        if col in df.columns:
            df[col] = df[col].astype("category")

    # 정수형 컬럼
    for col, dtype in [("year", "Int16"), ("month", "Int8"), ("rating", "Int8"), ("helpful_count", "Int32")]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").astype(dtype)

    # 0/1 플래그 컬럼
    for col in ["has_image", "women_size_yn"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0).astype("int8")

    # 실수형 컬럼
    for col in ["user_height", "user_weight"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").astype("float32")

    return df

# 사용
df = convert_dtypes(df)
df.info(memory_usage="deep")
```

```python
import re

def clean_text(text):
    # 1. 결측치 처리
    if pd.isna(text):
        return ' '
    text = str(text)
    
    # 2. HTML 태그 제거
    text = re.sub(r'<[^>]+>', ' ', text)
    
    # 3. URL 제거
    text = re.sub(r'http\S+|www\.\S+', ' ', text)
    
    # 4. 이메일 제거
    text = re.sub(r'\S+@\S+', ' ', text)
    
    # 5. 이모지 & 이모티콘 제거
    text = re.sub(r'[^\w\s\uAC00-\uD7A3\u3130-\u318F]', ' ', text)
    
    # \uAC00-\uD7A3 : 한글 완성형
    # \u3130-\u318F : ㅎ, ㅋ 같은 한글 자모음 (반복 제거 대상이라 유지)
    
    # 6. 줄바꿈, 탭 → 공백으로 변환
    text = re.sub(r'[\n\r\t]', ' ', text)
    
    # 7. 반복 문자 1개로 통일 (ㅎㅎㅎ→ㅎ, ㅋㅋㅋ→ㅋ, ~~~→~)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 8. 연속 공백 → 공백 1개로 통일
    text = re.sub(r'\s+', ' ', text)
    
    # 9. 대소문자 통합 (영문 소문자)
    text = text.lower()
    
    # 10. 앞뒤 공백 제거
    text = text.strip()
    
    return text

df['content'] = df['content'].apply(clean_text)
```

```python
# =========================================
# 띄어쓰기 제거 후 5글자 이하인 데이터 확인
# =========================================
short_mask = df['content'].str.replace(' ', '', regex=False).str.len() <= 5

print(f"5글자 이하 행 수: {short_mask.sum()}")
display(df[short_mask]['content'].value_counts())
```

```python
# =========================
# 5글자 이하 행 제거
# =========================
before = len(df)
df = df[~short_mask]
after = len(df)

print(f"제거 전: {before}개")
print(f"제거된 행: {before - after}개")
print(f"제거 후: {after}개")
```
