## 1. 전처리

```python
# 파일 불러오기
import pandas as pd
import numpy as np

df_review_fila_mss = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/reviews/musinsa_fila_shoes_reviews.csv")
display(df_review_fila_mss.head(3))
display(df_review_fila_mss.info())
```

```python
# =======================================
# 사용할 컬럼만 남기기
# =======================================
df_review_fila_mss = df_review_fila_mss.drop(['review_type', 'author', 'comment_count', 'image_urls', 'survey_size', 'survey_width', 'survey_comfort', 'survey_instep'], axis=1)
df_review_fila_mss.info() # 컬럼 13개
```

```python
# =======================================
# 결측치 처리
# =======================================
# 'review_text' => 14090개로 결측치 있음.

# 리뷰 텍스트 분석이 중요
df_review_fila_mss['review_text'].isna().mean()*100 # 0.07% => 행 제거로 결정

df_review_fila_mss = df_review_fila_mss.dropna(subset=['review_text'], axis=0)
df_review_fila_mss.info()
```

```python
# =======================================
# 중복된 데이터 제거
# =======================================
df_review_fila_mss = df_review_fila_mss.drop_duplicates(subset=[
        'review_date', 'product_id', 'product_name', 'review_no', 'rating', 'review_text', 'user_height', 'user_weight', 'option', 'helpful', 'has_image', 'url'
])

df_review_fila_mss.shape # (8575, 13)
```

```python
pd.set_option('display.max_colwidth', None)
print(df_review_fila_mss.duplicated(subset=['review_date', 'product_id', 'product_name', 'rating', 'review_text', 'user_height', 'user_weight', 'option']).sum())
display(df_review_fila_mss[df_review_fila_mss.duplicated(subset=['review_date', 'product_id', 'product_name', 'rating', 'review_text', 'user_height', 'user_weight', 'option'])].head())
```

```python
subset_cols = ['review_date', 'product_id', 'product_name', 'rating', 'review_text', 'user_height', 'user_weight', 'option']

# has_image 내림차순, helpful 내림차순 정렬 → 각 그룹에서 첫 번째 행만 남김
df_review_fila_mss = (
    df_review_fila_mss
    .sort_values(['has_image', 'helpful'], ascending=[False, False])
    .drop_duplicates(subset=subset_cols, keep='first')
    .sort_index()
)

df_review_fila_mss.shape # 8575 - 366 = 8209 (잘 지워진 것 확인 완료)
```

```python
df_review_fila_mss.info()
```

```python
# =======================================
# 결측치 처리
# =======================================
# option => 8개 결측치 있음
df_review_fila_mss['option'].isna().mean()*100 # 0.097%

# 에샤페 실버문(1XM02348H063) 상품에만 결측치 있음.
df_review_fila_mss[df_review_fila_mss['option'].isna()]
```

```python
# option이 null인 product_id 추출
null_product_ids = df_review_fila_mss[df_review_fila_mss['option'].isna()]['product_id'].unique()

# 해당 product_id들의 option 분포 확인
df_review_fila_mss[df_review_fila_mss['product_id'].isin(null_product_ids)][['product_id', 'product_name', 'option']].value_counts(dropna=False)
# =>>> option = 범주형 데이터 => 최빈값으로 대체 (Grey/White/White · 240) => 결측치가 8개로 아주 작은 숫자라 대체해도 데이터 왜곡 적을 것임.
```

```python
df_review_fila_mss.loc[df_review_fila_mss['option'].isna(), 'option'] = 'Grey/White/White · 240'

print("===== 결측치 확인 ======")
print(df_review_fila_mss.info())
print("===== option 값 유지 확인 ======")
print(df_review_fila_mss['option'].value_counts())
```

```python
# =======================================
# 컬럼명 변경
# =======================================

df_review_fila_mss = df_review_fila_mss.rename(columns={
    'review_no' : 'review_id',
    'review_text' : 'content',
    'option' : 'purchase_option',
    'helpful' : 'helpful_count',
    'url' : 'product_url'
})

df_review_fila_mss.head(3)
```

```python
# =======================================
# 없는 컬럼 추가
# =======================================

# 없는 컬럼 추가하기
required_columns = [
    'collect_date', 'platform', 'review_id', 'product_id', 'product_name',
    'review_date', 'year', 'month', 'content', 'rating', 'helpful_count',
    'has_image', 'purchase_option', 'user_size', 'user_height', 'user_weight',
    'user_height_group', 'user_weight_group', 'product_url'
]

for col in required_columns:
    if col not in df_review_fila_mss.columns:
        df_review_fila_mss[col] = None

# =======================================
# 컬럼 순서 정리
# =======================================
df_review_fila_mss = df_review_fila_mss[required_columns]

df_review_fila_mss.info()
```

```python
# =======================================
# year, month 다시 생성하기
# =======================================
# 날짜 컬럼 형변환
df_review_fila_mss['collect_date'] = pd.to_datetime(df_review_fila_mss['collect_date']).dt.normalize()
df_review_fila_mss['review_date'] = pd.to_datetime(df_review_fila_mss['review_date']).dt.normalize()

# 기존 year, month 제거
df_review_fila_mss = df_review_fila_mss.drop(['year', 'month'], axis=1)

# 파생컬럼 생성
df_review_fila_mss['year'] = df_review_fila_mss['review_date'].dt.year
df_review_fila_mss['month'] = df_review_fila_mss['review_date'].dt.month
```

```python
# =======================================
# platform = 무신사
# =======================================
df_review_fila_mss['platform'] = '무신사'
df_review_fila_mss['platform'].value_counts()
```

```python
# =======================================
# 고객 신체 정보 None으로 채우기
# =======================================
df_review_fila_mss[['user_height', 'user_weight', 'user_height_group', 'user_weight_group']] = (
    df_review_fila_mss[['user_height', 'user_weight', 'user_height_group', 'user_weight_group']].astype(object).fillna(None)
)
```

```python
# =======================================
# has_image => 0, 1로 변환하기
# =======================================
df_review_fila_mss['has_image'] = df_review_fila_mss['has_image'].map({
    True : 1,
    False : 0
})
df_review_fila_mss['has_image'].value_counts()
```

```python
df_review_fila_mss.info()
```

```python
# =======================================
# 데이터 타입 정리하기
# =======================================
df_review_fila_mss['platform'] = df_review_fila_mss['platform'].astype('category')
df_review_fila_mss['purchase_option'] = df_review_fila_mss['purchase_option'].astype('category')

df_review_fila_mss[['platform', 'purchase_option']].info()
```

```python
# =======================================
# 컬럼 순서 변경
# =======================================
df_review_fila_mss = df_review_fila_mss[['collect_date', 'platform', 'review_id', 'product_id', 'product_name',
    'review_date', 'year', 'month', 'content', 'rating', 'helpful_count',
    'has_image', 'purchase_option', 'user_size', 'user_height', 'user_weight',
    'user_height_group', 'user_weight_group', 'product_url']]

df_review_fila_mss.head(3)
```

```python
# =======================================
# 이상치 확인
# =======================================
min(df_review_fila_mss['review_date']) # 2024-02-27
max(df_review_fila_mss['review_date']) # 2026-04-13

df_review_fila_mss['platform'].value_counts() # 무신사

df_review_fila_mss['review_id'].value_counts(ascending=False) # 모든 값 1

df_review_fila_mss['product_id'].value_counts(ascending=False)

df_review_fila_mss['product_name'].value_counts(ascending=False)

min(df_review_fila_mss['year']) # 2024
max(df_review_fila_mss['year']) # 2026

min(df_review_fila_mss['month']) # 1
max(df_review_fila_mss['month']) # 12
```

```python
# =======================================
# 저장
# =======================================
df_review_fila_mss_2024 = df_review_fila_mss[df_review_fila_mss['year'] >= 2024]

df_review_fila_mss.to_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/final/fila_musinsa_review_final.csv", encoding='utf-8-sig', index=False)
df_review_fila_mss_2024.to_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/final/fila_musinsa_review_final_s2024.csv", encoding='utf-8-sig', index=False)
```
