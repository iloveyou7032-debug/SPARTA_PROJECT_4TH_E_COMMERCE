```python
# 필요한 컬럼 선택

df_review_fila_D2C = df_review_fila_D2C[[
    'review_id', 'collect_date', 'post_date', 'platform', 'search_keyword', 'content', 'source_url', 'rating', 'helpful_count', 'purchased_option', 'has_image'
]] # 11개

display(df_review_fila_D2C.head())
display(df_review_fila_D2C.info())
```

```python
# merge 전 product_id 매칭 확인하기
# 제품 리스트의 product_id
product_ids = set(df_product_fila_D2C['product_id'])

# 리뷰 데이터의 search_keyword
search_keywords = set(df_review_fila_D2C['search_keyword'])

# 매칭 확인
matched = product_ids & search_keywords
unmatched = search_keywords - product_ids

print(f"매칭되는 값: {len(matched)}개")
print(f"매칭 안 되는 search_keyword: {len(unmatched)}개")
print(unmatched)
```

```python
# 제품 데이터에서 product_name, raw_colors 추가
df_review_fila_D2C = df_review_fila_D2C.merge(
    df_product_fila_D2C[['product_id', 'product_name', 'raw_colors']],  # 필요한 컬럼만 선택
    left_on='search_keyword',
    right_on='product_id',
    how='left'  # 리뷰 데이터 기준으로 merge
)

df_review_fila_D2C.head(2)

# merge 확인
# df_review_fila_D2C['product_id'].isna().sum() =>>>>>> product_id => null 없음
# df_review_fila_D2C[df_review_fila_D2C['search_keyword'] != df_review_fila_D2C['product_id']] =>>>>>> 'search_keyword'와 'product_id'가 잘 매칭 됐는지 확인 완료!
# df_product_fila_D2C['raw_colors'].isna().sum() =>>>>>>>>>> raw_colors 원래 null 값 있었는지 확인 완료!
```

```python
# raw_colors + purchased_option
import numpy as np

df_review_fila_D2C['purchase_option_combined'] = np.where(
    df_review_fila_D2C['purchased_option'].isna(),
    
    # 1) purchased_option이 NaN이면 → raw_colors 그대로
    df_review_fila_D2C['raw_colors'],
    
    # 2) purchased_option이 있으면
    np.where(
        df_review_fila_D2C['raw_colors'].isna(),
        
        # raw_colors도 NaN이면 → purchased_option만
        df_review_fila_D2C['purchased_option'],
        
        # 둘 다 있으면 → 합치기
        df_review_fila_D2C['purchased_option'] + ' / 컬러:' + df_review_fila_D2C['raw_colors']
    )
)

pd.set_option('display.max_rows', None)
df_review_fila_D2C['purchase_option_combined'].value_counts(dropna=False)
```

# 휠라 무신사 리뷰 데이터

```python
# 파일 불러오기
import pandas as pd
import numpy as np

df_review_fila_mss = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/reviews/musinsa_fila_shoes_reviews.csv")
display(df_review_fila_mss.head(3))
display(df_review_fila_mss.info())
```

```python
# =======================================
# 사용할 컬럼만 남기기
# =======================================
df_review_fila_mss = df_review_fila_mss.drop(['review_type', 'author', 'comment_count', 'image_urls', 'survey_size', 'survey_width', 'survey_comfort', 'survey_instep'], axis=1)
df_review_fila_mss.info() # 컬럼 13개
```

```python
# =======================================
# 결측치 처리
# =======================================
# 'review_text' => 14090개로 결측치 있음.

# 리뷰 텍스트 분석이 중요
df_review_fila_mss['review_text'].isna().mean()*100 # 0.07% => 행 제거로 결정

df_review_fila_mss = df_review_fila_mss.dropna(subset=['review_text'], axis=0)
df_review_fila_mss.info()
```

```python
# =======================================
# 중복된 데이터 제거
# =======================================
df_review_fila_mss = df_review_fila_mss.drop_duplicates(subset=[
        'review_date', 'product_id', 'product_name', 'review_no', 'rating', 'review_text', 'user_height', 'user_weight', 'option', 'helpful', 'has_image', 'url'
])

df_review_fila_mss.shape # (8575, 13)
```

```python
pd.set_option('display.max_colwidth', None)
print(df_review_fila_mss.duplicated(subset=['review_date', 'product_id', 'product_name', 'rating', 'review_text', 'user_height', 'user_weight', 'option']).sum())
display(df_review_fila_mss[df_review_fila_mss.duplicated(subset=['review_date', 'product_id', 'product_name', 'rating', 'review_text', 'user_height', 'user_weight', 'option'])].head())
```
