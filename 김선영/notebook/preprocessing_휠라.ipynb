# 휠라 네이버스토어 리뷰 데이터

## 1. 전처리

```python
# 데이터 불러오기
df_review_fila_naverstore = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/reviews/FILA_NaverStore_Review_since2020.csv", encoding='utf-8')
display(df_review_fila_naverstore.head())
```

```python
# 필요한 컬럼 선택

df_review_fila_naverstore = df_review_fila_naverstore[[
    'product_url', 'product_id', 'review_id', 'review_body', 'rating', 'helpful_count', 'review_date', 'review_content_class', 'extra_productName', 'extra_productOptionContent'
]] # 10개

display(df_review_fila_naverstore.head())
display(df_review_fila_naverstore.info())
```

```python
# review_content_class 변환
df_review_fila_naverstore['review_content_class'].value_counts() # PHOTO, TEXT, VIDEO

# PHOTO, VIDEO = 1
# TEXT = 0
df_review_fila_naverstore['review_content_class'] = df_review_fila_naverstore['review_content_class'].map({
    'PHOTO' : 1,
    'TEXT' : 0,
    'VIDEO' : 1
})

print(df_review_fila_naverstore['review_content_class'].value_counts())
print(df_review_fila_naverstore.info())
```

```python
# 데이터 타입 변환
# rating, helpful_count, review_date
df_review_fila_naverstore['rating'] = df_review_fila_naverstore['rating'].astype(int)
df_review_fila_naverstore['helpful_count'] = df_review_fila_naverstore['helpful_count'].astype(int)
df_review_fila_naverstore['review_date'] = pd.to_datetime(df_review_fila_naverstore['review_date']).dt.normalize().dt.tz_localize(None)

print(df_review_fila_naverstore.info())
```

```python
# 파생컬럼 생성 -> year, month
df_review_fila_naverstore['year'] = df_review_fila_naverstore['review_date'].dt.year
df_review_fila_naverstore['month'] = df_review_fila_naverstore['review_date'].dt.month

df_review_fila_naverstore.head(3)
```

```python
df_review_fila_naverstore.info()
```

```python
# 컬럼명 정리
df_review_fila_naverstore = df_review_fila_naverstore.rename(columns={
    'review_content_class' : 'has_image',
    'extra_productName' : 'product_name',
    'extra_productOptionContent' : 'purchase_option'
})

df_review_fila_naverstore.head(1)
```

```python
# 없는 컬럼 추가하기
required_columns = [
    'collect_date', 'platform', 'review_id', 'product_id', 'product_name',
    'review_date', 'year', 'month', 'review_body', 'rating', 'helpful_count',
    'has_image', 'purchase_option', 'user_height', 'user_weight',
    'user_height_group', 'user_weight_group', 'product_url'
]

for col in required_columns:
    if col not in df_review_fila_naverstore.columns:
        df_review_fila_naverstore[col] = None

df_review_fila_naverstore.head()
```

```python
# 컬럼 순서 정리
df_review_fila_naverstore = df_review_fila_naverstore[required_columns]
df_review_fila_naverstore.head()
```

```python
df_review_fila_naverstore.info()
```

```python
# 이상치 확인
df_review_fila_naverstore['review_date'].value_counts(dropna=False)
```

```python
# review_id 중복 확인
df_review_fila_naverstore['review_id'].duplicated().sum()
```

```python
df_review_fila_mss = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/reviews/musinsa_fila_shoes_reviews.csv", encoding='utf-8')
display(df_review_fila_mss.head(3))
display(df_review_fila_mss.info())
```

```python
# 사용 컬럼 선정
df_review_fila_mss = df_review_fila_mss[['review_date', 'product_id', 'product_name', 'review_no', 'rating', 'review_text', 'url']]
display(df_review_fila_mss.head(3))
display(df_review_fila_mss.info())
```

```python
# 결측치 처리
# 리뷰 분석이 중요하기 때문에 review_text가 null인 컬럼은 제거
df_review_fila_mss = df_review_fila_mss.dropna(subset=['review_text'], axis=0)
df_review_fila_mss.info()
```

```python
# review_date 파생컬럼
df_review_fila_mss['review_date'] = pd.to_datetime(df_review_fila_mss['review_date'])

df_review_fila_mss['year'] = df_review_fila_mss['review_date'].dt.year
df_review_fila_mss['month'] = df_review_fila_mss['review_date'].dt.month

display(df_review_fila_mss.head())
display(df_review_fila_mss[['review_date', 'year', 'month']].info())
```

```python
# 컬럼명 변경
df_review_fila_mss = df_review_fila_mss.rename(columns={
    'review_no' : 'review_id',
    'review_text' : 'review_body',
    'url' : 'product_url'
})

df_review_fila_mss.head()
```

```python
# 컬럼순서 변경
df_review_fila_mss = df_review_fila_mss[['review_id', 'product_id', 'product_name', 'review_date', 'year', 'month', 'review_body', 'rating', 'product_url']]
df_review_fila_mss.head()
```
