# 휠라 네이버스토어 리뷰 데이터

## 1. 전처리

```python
# =============================
# 결측값 확인
# =============================

print("===== 전체 컬럼 결측치 확인 =====")
df_review_fila_naverstore.info()

# 1)
# 'Unnamed: 0' : 인덱스였던 데이터 같음 -> 필요 없으니 삭제

# 2)
# 'extra_reviewTopics' 결측치 있음. -> 필요없는 데이터라 컬럼 삭제
print("\n===== 'extra_reviewTopics' 결측치 확인(%) =====")
print(df_review_fila_naverstore['extra_reviewTopics'].isna().mean()*100)
```

```python
# 필요한 컬럼 선택
# 'product_url', 'product_id', 'review_id', 'review_body', 'rating', 'review_date', 'extra_productName'

# 컬럼 선택
df_review_fila_naverstore = df_review_fila_naverstore[['product_url', 'product_id', 'review_id', 'review_body', 'rating', 'review_date', 'extra_productName']]
df_review_fila_naverstore.head()
# dtype 통일
```

```python
# =============================
# 컬럼 명 통일
# =============================
df_review_fila_naverstore = df_review_fila_naverstore.rename(columns={
    'extra_productName' : 'product_name'
})
```

```python
# =============================
# dtype 변환
# =============================
# df_review_fila_naverstore.info()

# rating -> int
# 데이터 다 정수 맞음. -> int 변환 가능
df_review_fila_naverstore['rating'] = df_review_fila_naverstore['rating'].astype(int)

# review_date -> datetime
df_review_fila_naverstore['review_date'] = pd.to_datetime(df_review_fila_naverstore['review_date'])
# UTC -> KST(Asia/Seoul)
# df_review_fila_naverstore['review_date'] = df_review_fila_naverstore['review_date'].dt.tz_convert('Asia/Seoul')
# 소수점 초 없애기
df_review_fila_naverstore['review_date'] = df_review_fila_naverstore['review_date'].dt.strftime('%Y-%m-%d %H:%M:%S')
# # review_date -> datetime
df_review_fila_naverstore['review_date'] = pd.to_datetime(df_review_fila_naverstore['review_date'])

print("===== dtype 번환 후 info =====")
display(df_review_fila_naverstore.info())
display(df_review_fila_naverstore.head())
```

```python
# =============================
# 파생컬럼 생성 -> year, month
# =============================
df_review_fila_naverstore['year'] = df_review_fila_naverstore['review_date'].dt.year
df_review_fila_naverstore['month'] = df_review_fila_naverstore['review_date'].dt.month

display(df_review_fila_naverstore[['year', 'month']].info())
df_review_fila_naverstore.head()
```

```python
# =============================
# 컬럼 순서 통일
# =============================
df_review_fila_naverstore = df_review_fila_naverstore[['review_id', 'product_id', 'product_name', 'review_date', 'year', 'month', 'review_body', 'rating', 'product_url']]

print("========== 휠라 네이버스토어 리뷰 ==========")
display(df_review_fila_naverstore.head())
display(df_review_fila_naverstore.info())
```

```python
# =============================
# 이상값 확인
# =============================

df_review_fila_naverstore['product_url'].value_counts()
```

```python
df_review_fila_mss = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/reviews/musinsa_fila_shoes_reviews.csv", encoding='utf-8')
display(df_review_fila_mss.head(3))
display(df_review_fila_mss.info())
```

```python
# 사용 컬럼 선정
df_review_fila_mss = df_review_fila_mss[['review_date', 'product_id', 'product_name', 'review_no', 'rating', 'review_text', 'url']]
display(df_review_fila_mss.head(3))
display(df_review_fila_mss.info())
```

```python
# 결측치 처리
# 리뷰 분석이 중요하기 때문에 review_text가 null인 컬럼은 제거
df_review_fila_mss = df_review_fila_mss.dropna(subset=['review_text'], axis=0)
df_review_fila_mss.info()
```

```python
# 컬럼명 변경
df_review_fila_mss = df_review_fila_mss.rename(columns={
    'review_no' : 'review_id',
    'review_text' : 'review_body',
    'url' : 'product_url'
})

df_review_fila_mss.head()
```

```python
# 컬럼순서 변경
df_review_fila_mss = df_review_fila_mss[['review_id', 'product_id', 'product_name', 'review_date', 'year', 'month', 'review_body', 'rating', 'product_url']]
df_review_fila_mss.head()
```
