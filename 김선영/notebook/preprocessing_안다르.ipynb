```python
# 안다르
df_review_andar_D2C = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/reviews/andar_reviews_final_new.csv")
display(df_review_andar_D2C.head(3))
display(df_review_andar_D2C.info())
```

```python
# =======================================
# 사용할 컬럼만 선택
# =======================================
df_review_andar_D2C = df_review_andar_D2C.drop(['post_time', 'brand', 'user_nickname', 'media_count', 'main_content_type', 'upload_from', 'labels', 'topics', '평소사이즈', '사이즈_평가'], axis=1)
df_review_andar_D2C.head(3)
```

```python
# =======================================
# 리뷰 본문 합치기 => content + title, 사이즈_한줄평
# =======================================
# 합치기
df_review_andar_D2C['content'] = df_review_andar_D2C[['content', 'title', '사이즈_한줄평']].fillna('').agg(' '.join, axis=1).str.strip()

# 합친 컬럼 삭제하기
df_review_andar_D2C = df_review_andar_D2C.drop(['title', '사이즈_한줄평'], axis=1)

display(df_review_andar_D2C.head())
```

```python
# purchased_option => 18.8%
pd.set_option('display.max_colwidth', None) # 컬럼 안 텍스트 잘림 제거
df_review_andar_D2C[['search_keyword', 'source_url', 'purchased_option']].drop_duplicates().head(20)
# ‘product_id’와 제품 옵션이 1:1로 매칭되지 않음. 결측치 그대로 유지
```

```python
# 키 => 31.5%
df_review_andar_D2C['키'] = df_review_andar_D2C['키'].astype(object)
df_review_andar_D2C['키'] = df_review_andar_D2C['키'].fillna(None) # None으로 잘 대체 됨.

# 몸무게 => 32.4%
df_review_andar_D2C['몸무게'] = df_review_andar_D2C['몸무게'].astype(object)
df_review_andar_D2C['몸무게'] = df_review_andar_D2C['몸무게'].fillna(None) # None으로 잘 대체 됨.

# display(df_review_andar_D2C['키'].value_counts(dropna=False))
# display(df_review_andar_D2C['몸무게'].value_counts(dropna=False))
```

```python
# =======================================
# 파생컬럼 생성 (year, month)
# =======================================
df_review_andar_D2C['post_date'] = pd.to_datetime(df_review_andar_D2C['post_date'])

df_review_andar_D2C['year'] = df_review_andar_D2C['post_date'].dt.year
df_review_andar_D2C['month'] = df_review_andar_D2C['post_date'].dt.month
```

```python
# =======================================
# 컬럼 이름 통일
# =======================================

df_review_andar_D2C = df_review_andar_D2C.rename(columns={
    'post_date' : 'review_date', 
    'search_keyword' : 'product_name',
    'content' : 'review_body',
    'source_url' : 'product_url',
    'purchased_option' : 'purchase_option',
    '키' : 'user_height',
    '몸무게' : 'user_weight'
})
```

```python
# =======================================
# 없는 컬럼 추가
# =======================================

required_columns = [
    'collect_date', 'platform', 'review_id', 'product_id', 'product_name',
    'review_date', 'year', 'month', 'review_body', 'rating', 'helpful_count',
    'has_image', 'purchase_option', 'user_height', 'user_weight',
    'user_height_group', 'user_weight_group', 'product_url'
]

for col in required_columns:
    if col not in df_review_andar_D2C.columns:
        df_review_andar_D2C[col] = None

# 컬럼 순서 정리
df_review_andar_D2C = df_review_andar_D2C[required_columns]
```

```python
df_review_andar_D2C.info()
```

```python
# =======================================
# 데이터 타입 통일
# =======================================

df_review_andar_D2C['collect_date'] = pd.to_datetime(df_review_andar_D2C['collect_date'])
df_review_andar_D2C['platform'] = df_review_andar_D2C['collect_date'].astype('category')
df_review_andar_D2C['purchase_option'] = df_review_andar_D2C['purchase_option'].astype('category')
# df_review_andar_D2C['purchase_option'] = df_review_andar_D2C['purchase_option'].astype('category')
```

```python
# =======================================
# 데이터 값 정리
# =======================================

# platform 데이터 => 자사몰
df_review_andar_D2C['platform'] = '자사몰'

# purchase_option => None
df_review_andar_D2C['purchase_option'] = df_review_andar_D2C['purchase_option'].astype(object)
df_review_andar_D2C['purchase_option'] = df_review_andar_D2C['purchase_option'].fillna(None)

# 신체정보 컬럼이름 바꾸기, 순서 변경하기
df_review_andar_D2C = df_review_andar_D2C.rename(columns={
    'user_height': 'user_height_temp',
    'user_weight': 'user_weight_temp',
    'user_height_group': 'user_height',
    'user_weight_group': 'user_weight',
    'user_height_temp': 'user_height_group',
    'user_weight_temp': 'user_weight_group',
})
```

```python
# 신체정보 컬럼이름 바꾸기, 순서 변경하기
df_review_andar_D2C = df_review_andar_D2C.rename(columns={
    'user_height_temp': 'user_height',
    'user_weight_temp': 'user_weight',
})

df_review_andar_D2C = df_review_andar_D2C[required_columns]
```

```python
# =======================================
# 데이터 타입 통일
# =======================================

df_review_andar_D2C['user_height_group'] = df_review_andar_D2C['user_height_group'].astype('category')
df_review_andar_D2C['user_weight_group'] = df_review_andar_D2C['user_weight_group'].astype('category')
```

```python
# =======================================
# user_height_group, user_weight_group => 데이터 확인
# =======================================
display(df_review_andar_D2C['user_height_group'].value_counts())
display(df_review_andar_D2C['user_weight_group'].value_counts())
```

```python
# ====================================================================================
# purchase_option, user_height_group, user_weight_group => fillna(None)
# ====================================================================================
df_review_andar_D2C[['purchase_option', 'user_height_group', 'user_weight_group']] = (
    df_review_andar_D2C[['purchase_option', 'user_height_group', 'user_weight_group']]
    .astype(object)
    .fillna(None)
)
```

```python
# =======================================
# 'is_outlet' 제거
# =======================================

df_review_andar_D2C = df_review_andar_D2C.drop(['is_outlet'], axis=1)
```

```python
# =======================================
# dtype
# =======================================

df_review_andar_D2C['platform'] = df_review_andar_D2C['platform'].astype('category')
df_review_andar_D2C['purchase_option'] = df_review_andar_D2C['purchase_option'].astype('category')
df_review_andar_D2C['user_height_group'] = df_review_andar_D2C['user_height_group'].astype('category')
df_review_andar_D2C['user_weight_group'] = df_review_andar_D2C['user_weight_group'].astype('category')
```

### 중복제거

```python
df_review_andar_D2C_copy = df_review_andar_D2C.copy()
```

```python
# =================================================
# product_id, product_name, review_body (완료)
# =================================================

# 1) 중복 데이터 확인

df_review_andar_D2C[df_review_andar_D2C.duplicated(subset=['product_id', 'product_name', 'review_body'], keep=False)]\
    .sort_values(['product_id', 'review_body'])[['product_id', 'product_name', 'review_body']].head(10)
```

```python
# 3) 중복제거 확인
df_review_andar_D2C_copy.duplicated(
    subset=['product_id', 'product_name', 'review_body']
).sum()
```

```python
# =========================================
# product_name & product_id 매창, 중복 확인 (하는 중이었음)
# =========================================

# 1) product_id, product_name 1:1 매칭 확인
# product_id에 2개 매칭된 거 데이터 확인
df_review_andar_D2C.groupby('product_id')['product_name'].nunique().sort_values(ascending=False)
```
