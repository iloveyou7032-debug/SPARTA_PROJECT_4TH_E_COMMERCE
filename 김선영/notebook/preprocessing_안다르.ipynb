```python
import pandas as pd
import numpy as np

df_review_andar_mss = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/reviews/musinsa_andar_reviews.csv", encoding='utf-8')
display(df_review_andar_mss.head())
display(df_review_andar_mss.info())
```

```python
# ======================
# 사용할 컬럼만 남기기
# ======================
df_review_andar_mss = df_review_andar_mss.drop(['review_type', 'author', 'comment_count', 'image_urls', 'survey_json'], axis=1, errors='ignore')
df_review_andar_mss.head(3) # (7552, 13)
```

```python
display(df_review_andar_mss.head(1))
display(df_review_andar_mss.info())
```

```python
# ======================
# 데이터 타입 통일
# ======================
df_review_andar_mss['collect_date'] = pd.to_datetime(df_review_andar_mss['collect_date'])
df_review_andar_mss['review_date'] = pd.to_datetime(df_review_andar_mss['review_date'])
df_review_andar_mss['product_id'] = df_review_andar_mss['product_id'].astype(int)
df_review_andar_mss['has_image'] = df_review_andar_mss['has_image'].astype(int)
df_review_andar_mss['option'] = df_review_andar_mss['option'].astype('category')
```

```python
# ===========================
# 파생 컬럼 생성 (year, month)
# ===========================
df_review_andar_mss['year'] = df_review_andar_mss['review_date'].dt.year
df_review_andar_mss['month'] = df_review_andar_mss['review_date'].dt.month
```

```python
# ===========================
# option 결측치 처리 (0.079%)
# ===========================

# option 컬럼의 결측치 비율 확인
df_review_andar_mss['option'].isna().mean()*100

# 실제 option 컬럼의 값 확인 (6개 행)
pd.set_option('display.max_rows', None) # 행 보이기
df_review_andar_mss[df_review_andar_mss['option'].isna()] #.value_counts(dropna=False)

# >>>>> 제품 아이디 & 옵션 1:1 매칭이 아니라 대체값 없이 결측 유지 <<<<<
```

```python
# ===========================
# 컬럼 이름 변환
# ===========================
df_review_andar_mss = df_review_andar_mss.rename(columns={
    'review_no': 'review_id',
    'review_text' : 'content',
    'option' : 'purchase_option',
    'helpful' : 'helpful_count',
    'url' : 'product_url'
})

# 컬럼명 잘 변환된 거 확인 완료!
# Index(['collect_date', 'review_date', 'product_id', 'product_name',
#        'review_id', 'rating', 'review_body', 'user_height', 'user_weight',
#        'purchase_option', 'helpful_count', 'has_image', 'product_url', 'year',
#        'month'],
#       dtype='str')
```

```python
# ===========================
# 컬럼 추가
# ===========================
required_columns = [
    'collect_date', 'platform', 'review_id', 'product_id', 'product_name',
    'review_date', 'year', 'month', 'content', 'rating', 'helpful_count',
    'has_image', 'purchase_option', 'user_size', 'user_height', 'user_weight',
    'user_height_group', 'user_weight_group', 'product_url'
]

for col in required_columns:
    if col not in df_review_andar_mss.columns:
        df_review_andar_mss[col] = None

# ===========================
# 컬럼 순서 정리
# ===========================
df_review_andar_mss = df_review_andar_mss[required_columns]
```

```python
# ===========================
# 데이터 값 확인
# ===========================
df_review_andar_mss['collect_date'].value_counts(dropna=False)
df_review_andar_mss['platform'].value_counts(dropna=False)
df_review_andar_mss['user_height_group'].value_counts(dropna=False)
df_review_andar_mss['user_weight_group'].value_counts(dropna=False)
df_review_andar_mss['product_url'].value_counts(dropna=False).sort_values(ascending=False).head(10)

# collect_date 시간 제거해야 함
# platform 데이터 => 무신사 
# user_height_group => 세그먼트 해야 함
# user_weight_group => 세그먼트 해야 함
```

```python
display(df_review_andar_mss.head(1))
display(df_review_andar_mss.info())
```

```python
# 1) collect_date 시간 제거
df_review_andar_mss['collect_date'] = pd.to_datetime(df_review_andar_mss['collect_date']) # dtype('<M8[us]')
df_review_andar_mss['collect_date'] = df_review_andar_mss['collect_date'].dt.normalize() # 시간 제거
```

```python
# 2) platform 데이터 => 무신사
df_review_andar_mss['platform'] = '무신사' # 확인 완료
```

```python
# 3-1) user_height, user_weight == 0 => NaN으로 변경
df_review_andar_mss['user_height'] = df_review_andar_mss['user_height'].replace(0, np.nan)
df_review_andar_mss['user_weight'] = df_review_andar_mss['user_weight'].replace(0, np.nan)
```

```python
# 3) user_height_group 세그먼트
# 왼 미포함, 오른 포함
height_bins = [0, 139, 144, 149, 154, 159, 164, 169, 174, 179, 184, 189, 999]
height_labels = [
    '139cm 이하',
    '140~144cm',
    '145~149cm',
    '150~154cm',
    '155~159cm',
    '160~164cm',
    '165~169cm',
    '170~174cm',
    '175~179cm',
    '180~184cm',
    '185~189cm',
    '190cm 이상'
]
df_review_andar_mss['user_height_group'] = pd.cut(df_review_andar_mss['user_height'], bins=height_bins, labels=height_labels, right=True)

# 데이터 확인 완료
```

```python
# user_height_group => NaN 데이터 생긴 이유 확인
a = df_review_andar_mss[df_review_andar_mss['user_height_group'].isna()][['user_height', 'user_height_group']]
a.value_counts(dropna=False)

# >>>>> 'user_height', 'user_weight' 데이터가 없어서 NaN이 생김 = 정상 <<<<<
```

```python
# 4) user_weight_group 세그먼트
weight_bins = [0, 44, 49, 54, 59, 64, 69, 74, 79, 84, 89, 999]
weight_labels = [
    '44kg 이하',
    '45~49kg',
    '50~54kg',
    '55~59kg',
    '60~64kg',
    '65~69kg',
    '70~74kg',
    '75~79kg',
    '80~84kg',
    '85~89kg',
    '90kg 이상'
]
df_review_andar_mss['user_weight_group'] = pd.cut(df_review_andar_mss['user_weight'], bins=weight_bins, labels=weight_labels, right=True)

# 데이터 확인 완료
```

```python
# user_weight_group => NaN 데이터 생긴 이유 확인
b = df_review_andar_mss[df_review_andar_mss['user_weight_group'].isna()][['user_weight', 'user_weight_group']]
b.value_counts(dropna=False)

# >>>>> 'user_height', 'user_weight' 데이터가 없어서 NaN이 생김 = 정상 <<<<<
```

```python
display(df_review_andar_mss.head(1))
display(df_review_andar_mss.info())
```

```python
# ======================
# 데이터 타입 통일
# ======================
df_review_andar_mss['platform'] = df_review_andar_mss['platform'].astype('category')

df_review_andar_mss['user_weight'] = pd.to_numeric(df_review_andar_mss['user_weight'], errors='coerce').astype('Int64')
df_review_andar_mss['user_height'] = pd.to_numeric(df_review_andar_mss['user_height'], errors='coerce').astype('Int64')
```
