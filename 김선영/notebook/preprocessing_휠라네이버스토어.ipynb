```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/kimsunyeong/Desktop/SPARTA_PROJECT_4TH_E_COMMERCE/김선영/data/fila_naver_store_reviews/reviews_since_2020.csv")
df.head()
```

```python
df.info()
```

```python
# ================================================================================================================
# 사용할 컬럼 선정
# 'product_url', 'product_id', 'review_id', 'review_body', 'rating', 'helpful_count', 'review_date', 'reviewer_id', 'review_content_class'
# 'extra_id', 'extra_reviewContentClassType', 'extra_reviewScore', 'extra_reviewContent', 'extra_createDate', 'extra_writerId', 'extra_productNo', 'extra_productName', 'extra_productOptionContent', 'extra_reviewTopics', 
#   ㄴ리뷰ID
# ================================================================================================================

df2 = df[['product_url', 'product_id', 'review_id', 'review_body', 'rating', 'helpful_count', 'review_date', 'reviewer_id', 'review_content_class',
        'extra_id', 'extra_reviewContentClassType', 'extra_reviewScore', 'extra_reviewContent', 'extra_createDate', 'extra_writerId', 'extra_productNo',
        'extra_productName', 'extra_productOptionContent', 'extra_reviewTopics']
]

display(df2.head())
print(df2.info())
```

```python
# ================================================================================================================
# dtype 정리
# ================================================================================================================

df2['review_date'] = pd.to_datetime(df2['review_date'])
df2['extra_createDate'] = pd.to_datetime(df2['extra_createDate'])

df2[['review_date', 'extra_createDate']].info()
```

```python
# ================================================================================================================
# review_body 있는 것들만 남기기
# ================================================================================================================

df_body = df2.dropna(subset=['review_body'], axis=0)
df_body.info()
```

```python
# ================================================================================================================
# 컬럼 매칭 확인
# ================================================================================================================

# 1. product_id = extra_productNo
print("product_id != extra_productNo 개수: ", (df_body['product_id'] != df_body['extra_productNo']).sum(), "개")
# 1277개 : 전체 데이터가 매칭되지 않음. -> url에 표시되는 건 'product_id'이 맞음 -> 이걸로 결정

# 2. review_id = extra_id
print("review_id != extra_id 개수: ", (df_body['review_id'] != df_body['extra_id']).sum(), "개")

# 3. review_body = extra_reviewContent
print("review_body != extra_reviewContent 개수: ", (df_body['review_body'] != df_body['extra_reviewContent']).sum(), "개")

# 4. review_date = extra_createDate
print("review_date != extra_createDate 개수: ", (df_body['review_date'] != df_body['extra_createDate']).sum(), "개")

# 5. reviewer_id = extra_writerId
print("reviewer_id != extra_writerId 개수: ", (df_body['reviewer_id'] != df_body['extra_writerId']).sum(), "개")

# 6. review_content_class = extra_reviewContentClassType
print("review_content_class != extra_reviewContentClassType 개수: ", (df_body['review_content_class'] != df_body['extra_reviewContentClassType']).sum(), "개")
```
