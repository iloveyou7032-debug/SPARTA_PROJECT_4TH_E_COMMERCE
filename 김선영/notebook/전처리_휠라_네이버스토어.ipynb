```python
import re

def clean_text(text):
    # 1. 결측치 처리
    if pd.isna(text):
        return ' '

    text = str(text)

    # 2. HTML 태그 제거
    text = re.sub(r'<[^>]+>', ' ', text)

    # 3. URL 제거
    text = re.sub(r'http\S+|www\.\S+', ' ', text)

    # 4. 이메일 제거
    text = re.sub(r'\S+@\S+', ' ', text)

    # 5. 이모지 & 이모티콘 제거
    text = re.sub(r'[^\w\s\uAC00-\uD7A3\u3130-\u318F]', ' ', text)
    # \uAC00-\uD7A3 : 한글 완성형
    # \u3130-\u318F : ㅎ, ㅋ 같은 한글 자모음 (반복 제거 대상이라 유지)

    # 6. 줄바꿈, 탭 → 공백으로 변환
    text = re.sub(r'[\n\r\t]', ' ', text)

    # 7. 반복 문자 1개로 통일 (ㅎㅎㅎ→ㅎ, ㅋㅋㅋ→ㅋ, ~~~→~)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 8. 연속 공백 → 공백 1개로 통일
    text = re.sub(r'\s+', ' ', text)

    # 9. 대소문자 통합 (영문 소문자)
    text = text.lower()

    # 10. 앞뒤 공백 제거
    text = text.strip()

    return text

df_review_fila_naverstore_copy['content'] = df_review_fila_naverstore_copy['content'].apply(clean_text)
```

```python
# ========================================
# purchase_option 텍스트 정제 - 자동화 코드
# ========================================
import re

def normalize_purchase_option(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    
    # 1. 영문 소문자 통일
    text = text.lower()
    
    # 2. 콜론 뒤 띄어쓰기 제거 (색상: black → 색상:black)
    text = re.sub(r':\s+', ':', text)
    
    # 3. 슬래시 앞뒤 공백 통일 ( / )
    text = re.sub(r'\s*/\s*', ' / ', text)
    
    # 4. 앞뒤 공백 제거
    text = text.strip()
    
    return text

df_review_fila_naverstore_copy['purchase_option'] = (
    df_review_fila_naverstore_copy['purchase_option']
    .astype(str)
    .apply(normalize_purchase_option)
)
```

```python
# ======================================================================
# purchase_option 텍스트 정제 - 컬러 부분 슬래시( / ) → 쉼표(, )로 통일
# ======================================================================

import re

def fix_color_slash(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    
    def replace_with_size(m):
        color_part = re.sub(r'\s*/\s*', ',', m.group(1))
        return f'색상:{color_part} / {m.group(2)}'  # / 앞뒤 공백 유지
    
    def replace_without_size(m):
        color_part = re.sub(r'\s*/\s*', ',', m.group(1))
        return f'색상:{color_part}'
    
    if '사이즈:' in text:
        text = re.sub(r'색상:(.+?)\s*/\s*(사이즈:.+)', replace_with_size, text)
    else:
        text = re.sub(r'색상:(.+)', replace_without_size, text)
    
    return text

df_review_fila_naverstore_copy['purchase_option'] = df_review_fila_naverstore_copy['purchase_option'].apply(fix_color_slash)
```

```python
# ================================================
# purchase_option 텍스트 정제 - 색상 영문 → 한글 변환
# ================================================
color_map = {
    'dark navy': '다크네이비',
    'dark blue': '다크블루',
    'sky blue': '스카이블루',
    'light blue': '라이트블루',
    'off white': '오프화이트',
    'light pink': '라이트핑크',
    'dark pink': '다크핑크',
    'dark red': '다크레드',
    'dark gray': '다크그레이',
    'dark grey': '다크그레이',
    'melange gray': '멜란지그레이',
    'sand white': '샌드화이트',
    'light beige': '라이트베이지',
    'oatmeal melange': '오트밀멜란지',
    'black': '블랙',
    'brown': '브라운',
    'beige': '베이지',
    'white': '화이트',
    'cream': '크림',
    'blue': '블루',
    'navy': '네이비',
    'pink': '핑크',
    'red': '레드',
    'wine': '와인',
    'green': '그린',
    'khaki': '카키',
    'gray': '그레이',
    'grey': '그레이',
    'ivory': '아이보리',
    'purple': '퍼플',
    'yellow': '옐로우',
    'silver': '실버',
    'camel': '카멜',
    'orange' : '오렌지',
    'dark brown' : '다크브라운',
    '다크레드' : '다크레드'
}

def replace_colors(text):
    if pd.isna(text):
        return text
    
    # 색상: 이후 / 사이즈: 이전 부분만 추출해서 변환
    def convert_color_part(m):
        color_part = m.group(1)
        for eng, kor in color_map.items():
            # 단어 단위로 정확히 매칭 (앞뒤가 단어 경계)
            color_part = re.sub(r'\b' + eng + r'\b', kor, color_part)
        return f'색상:{color_part}'
    
    text = re.sub(r'색상:(.+?)(?= / 사이즈:|$)', convert_color_part, text)
    return text

df_review_fila_naverstore_copy['purchase_option'] = (
    df_review_fila_naverstore_copy['purchase_option']
    .apply(replace_colors)
)
```

```python
# ========================================
# purchase_option 텍스트 정제 - 수동
# ========================================

df_review_fila_naverstore_copy['purchase_option'] = df_review_fila_naverstore_copy['purchase_option'].replace({
    '색상:234(다크브라운) / 사이즈:85' : '색상:다크브라운 / 사이즈:85'
})
```

```python
# ========================================
# 여성 사이즈 여부 컬럼 생성
# 여자 사이즈는 '사이즈:' 뒤에 w가 있는 경우로 판단
# 1이면 여성 사이즈, 0이면 여성 사이즈 아님
# ========================================
df_review_fila_naverstore_copy['women_size_yn'] = (
    df_review_fila_naverstore_copy['purchase_option']
    .str.contains(r'사이즈:.*w', case=False, na=False)
    .astype(int)
) # 잘 됨 확인.
```

```python
# ========================================
# purchase_option 텍스트 정제 - 사이즈 텍스트 매핑
# ========================================
size_map = {
    '2xl': '110',
    'xl': '105',   # 필요시 추가
    'l': '100',    # 필요시 추가
    'm': '95',     # 필요시 추가
    's': '90',     # 필요시 추가
}

def normalize_size(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    
    def extract_size(m):
        size_val = m.group(1).lower()
        
        # 1. 사이즈 텍스트 매핑 먼저 처리 (2xl → 110 등)
        for key, val in size_map.items():
            if size_val == key:
                return f'사이즈:{val}'
        
        # 2. 숫자 추출
        numbers = re.findall(r'\d+', size_val)
        if numbers:
            return f'사이즈:{numbers[0]}'
        else:
            return f'사이즈:{m.group(1)}'  # 숫자 없으면 그대로 유지
    
    text = re.sub(r'사이즈:(\S+)', extract_size, text)
    return text

df_review_fila_naverstore_copy['purchase_option'] = (
    df_review_fila_naverstore_copy['purchase_option']
    .apply(normalize_size)
)
# tests = [
#     '색상:다크_네이비 / 사이즈:w70',
#     '색상:화이트,아이보리 / 사이즈:w95(wl)',
#     '색상:234(다크브라운) / 사이즈:ws(w85)',
#     '색상:브라운 / 사이즈:w85(ws)',
#     '색상:블루,라이트_블루 / 사이즈:w90(wm)',
#     '색상:블랙 / 사이즈:000',      # 숫자만 있는 경우 → 그대로
#     '색상:블랙 / 사이즈:free',     # 숫자 없는 경우 → 그대로
# ]
# for t in tests:
#     print(normalize_size(t))
```

```python
# ====================
# 추가 영어 매핑
# ====================
extra_color_map = {
    'dark 브라운': '다크브라운',
    'orange': '오렌지',
}

def clean_remaining(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    
    # 1. 색상 부분만 추출해서 처리
    def process_color(m):
        color_part = m.group(1)
        
        # 2. '선택XX.코드(한글명)' → 한글명만 추출
        color_part = re.sub(r'선택\d+\.\d+\([^)]+\)', lambda x: re.search(r'\(([^)]+)\)', x.group()).group(1), color_part)
        
        # 3. '숫자(한글명)' → 한글명만 추출
        color_part = re.sub(r'\d+\(([^)]+)\)', r'\1', color_part)
        
        # 4. '한글명(숫자)' → 한글명만 추출
        color_part = re.sub(r'([가-힣]+)\(\d+\)', r'\1', color_part)
        
        # 5. 남은 영어 한글 혼재 처리
        for eng, kor in extra_color_map.items():
            color_part = re.sub(r'\b' + eng + r'\b', kor, color_part, flags=re.IGNORECASE)
        
        return f'색상:{color_part}'
    
    text = re.sub(r'색상:(.+?)(?= / 사이즈:|$)', process_color, text)
    return text

df_review_fila_naverstore_copy['purchase_option'] = (
    df_review_fila_naverstore_copy['purchase_option']
    .apply(clean_remaining)
)
```

```python
# ========================================
# purchase_option 텍스트 정제 - 사이즈 앞의 0 제거
# ========================================
def remove_leading_zero_size(text):
    if pd.isna(text):
        return text
    
    text = str(text)
    
    # 사이즈: 뒤의 0으로 시작하는 숫자 → 앞의 0 제거 (000은 유지)
    text = re.sub(r'(사이즈:)0+([1-9]\d*)', r'\g<1>\2', text)
    
    return text

df_review_fila_naverstore_copy['purchase_option'] = (
    df_review_fila_naverstore_copy['purchase_option']
    .apply(remove_leading_zero_size)
)
```

```python
# ========================
# purchase_option 컬럼 분리
# ========================
def split_purchase_option(text):
    color, size = None, None
    
    if pd.isna(text):
        return pd.Series([None, None])
    
    parts = str(text).split(' / ')
    for part in parts:
        if part.startswith('색상:'):
            color = part.replace('색상:', '')
        elif part.startswith('사이즈:'):
            size = part.replace('사이즈:', '')
    
    return pd.Series([color, size])

df_review_fila_naverstore_copy[['purchase_option_color', 'purchase_option_size']] = (
    df_review_fila_naverstore_copy['purchase_option']
    .apply(split_purchase_option)
)

df_review_fila_naverstore_copy.head(10)
```

```python
# ========================
# 최종 컬럼 정리
# ========================

# 1) 데이터 형변환
def convert_dtypes(df):
    
    # 날짜 컬럼
    for col in ["collect_date", "review_date"]:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    # ID 컬럼
    for col in ["review_id", "product_id"]:
        if col in df.columns:
            df[col] = df[col].astype("string")

    # 문자열 컬럼
    for col in ["product_name", "content", "purchase_option", "product_url"]:
        if col in df.columns:
            df[col] = df[col].astype("string")

    # category 컬럼
    for col in ["platform", "purchase_option_color", "purchase_option_size", "user_height_group", "user_weight_group"]:
        if col in df.columns:
            df[col] = df[col].astype("category")

    # 정수형 컬럼
    for col, dtype in [("year", "Int16"), ("month", "Int8"), ("rating", "Int8"), ("helpful_count", "Int32")]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").astype(dtype)

    # 0/1 플래그 컬럼
    for col in ["has_image", "women_size_yn"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0).astype("int8")

    # 실수형 컬럼
    for col in ["user_height", "user_weight"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").astype("float32")

    return df

# 사용
df_review_fila_naverstore_copy = convert_dtypes(df_review_fila_naverstore_copy)
df_review_fila_naverstore_copy.info(memory_usage="deep")
```
